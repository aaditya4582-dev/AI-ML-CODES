```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset 1
data = {
    'Type of House': ['Farmhouse', 'Flat', 'Bungalow', 'Hostel', 'PG', 'Flat', 'Bungalow', 'Hostel', 'PG', 'Flat'],
    'Area of House': [100, 62, 110, 37, 25, 12, 5, 25, 12, 21],
    'House Size': ['1BHK', '2BHK', '4BHK', '1BHK', '1BHK', '2BHK', '3BHK', '2BHK', '3BHK', '2BHK'],
    'Cost': [100, 11, 66, 0, 77, 44, 55, 11, 16, 44]
}

df = pd.DataFrame(data)

print("--- Maximum Values in Dataset ---")
print(df.max())

print("\n--- Minimum Values in Dataset ---")
print(df.min())

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x="Area of House", y="Cost", hue="Type of House", data=df)

plt.xlabel("Area of house")
plt.ylabel("Cost")
plt.title("Area vs Cost by House Type")

plt.show()
```
