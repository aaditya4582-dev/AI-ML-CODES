```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

d = {
    "Customer ID": np.array([1000, 1021, 1022, 1037, 9870]),
    "Customer Name": np.array(['Shiv', 'Piyush', 'Yash', 'Arav', 'Deep']),
    "Product": np.array(['Protein bar', 'Maggi', 'Milk', 'Bread', 'Eggs']),
    "quantity": np.array([2, 5, 6, np.nan, 3]),
    "Cost": np.array([500, 50, 300, 75, 800]),
    "Category": np.array(['Health care', 'grocery', 'dairy', 'grocery', 'grocery'])
}

df = pd.DataFrame(d)
print("Initial DataFrame:")
print(df)

quan = df['quantity'].sum()
cost = df['Cost'].sum()

print(f"\nTotal Quantity: {quan}")
print(f"Total Cost: {cost}")

df['quantity'] = df['quantity'].fillna(df['quantity'].mean())

plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Cost', data=df)
plt.title("Cost by Category")
plt.show()

plt.figure(figsize=(8, 5))
sns.lineplot(x='Category', y='Cost', data=df, marker='o')
plt.title("Trend of Cost across Categories")
plt.show()
```
