```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = {
    'age': [20, 25, 30, 35, 40, 45],
    'height': [150, 160, 165, 170, 175, 180],
    'name': ['A', 'B', 'C', 'D', 'E', 'F']
}
df = pd.DataFrame(data)

sns.scatterplot(x='age', y='height', data=df, hue='name')
plt.title("Seaborn Scatterplot")
plt.show()

sns.displot(x='height', data=df)
plt.title("Height Distribution")
plt.show()

d1 = np.array([10, 20, 30, 40]) 
plt.pie(d1, labels=['Group 1', 'Group 2', 'Group 3', 'Group 4'], autopct='%1.1f%%')
plt.title("Data Distribution Pie Chart")
plt.show()
```
