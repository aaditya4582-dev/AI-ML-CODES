```python
graph = {}

a = int(input("Enter how many nodes we want: "))

for i in range(a):
 
    temp = input(f"Enter value of {i} node: ")
    graph[temp] = []
    
    b = int(input(f"Tell its no of neighbours for {temp}: "))
    
    for j in range(b):

        temp1 = input(f"Value of its {j} neighbor: ")
        graph[temp].append(temp1)

print("\nFinal Constructed Graph:")
print(graph)
```
