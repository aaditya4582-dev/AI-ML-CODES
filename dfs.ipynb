```python
def dfs(st, goal, graph, vis=None):
    if vis is None:
        vis = set()
    
    print(f"Visiting: {st}")
    
    if st == goal:
        return True
    
    vis.add(st)
    
    for child in graph[st]:
        if child not in vis:
            if dfs(child, goal, graph, vis):
                return True
                
    return False

graph1 = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'D'],
    'C': ['A'],
    'D': ['A', 'B']
}

print("Search A to C:")
print("Result:", dfs('A', 'C', graph1))

print("\nSearch A to D:")
print("Result:", dfs('A', 'D', graph1))
```
